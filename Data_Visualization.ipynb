```python
#importing libraries

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#line plot

x_values = ["Cakes","Bun","Coffee","Cookies","Milk"]
y_values = [5,10,35,22,50]

plt.plot(x_values,y_values)
```

```python
#bar plot

x_values = ["Cakes","Bun","Coffee","Cookies","Milk"]
y_values = [5,10,35,22,50]


plt.figure(figsize=(2,2))
plt.bar(x = x_values,
        height = y_values,
        width = 0.5,
        color="b")
plt.xticks(rotation=45,size=10)
```

```python
#using tencent data

df = pd.read_csv("tencent_data.csv")
```

```python
df
```

```python
#draw a bar plot of "Genre" vs "number of games available for that Genre"

df2 = (df.groupby("Genre")["Name"]
.count()
.reset_index()
.rename(columns={"Name":"Count"})
)
df2
```

```python
#plotting df2
x_bar = df2["Genre"]
y_bar = df2["Count"]

plt.figure(figsize=(6,4),facecolor="g")
plt.bar(x= x_bar,
        height= y_bar,
        width= 0.8,
        color = "r")
plt.xticks(rotation=45)
plt.title("Genre wise count of Tencent games")
plt.xlabel("Genre")
plt.ylabel("Count of games")
plt.show()
```

```python
# Showing Genre vs count using seaborn "sns"

sns.countplot(x= df["Genre"])
plt.xticks(rotation=45)
```

```python
#seaborn has some inbuilt datasets which they allow us to use

sns.get_dataset_names()
```

```python
#loading one of the inbuilt sns datasets

t = sns.load_dataset('titanic')
t
```

```python
#countplot of "class"

sns.countplot( data = t, x = 'class')
```

```python
#creating a dataframe t2 for barplot

t2 = t.groupby('class')[['class']].count().rename(columns = {'class': 'count'}).reset_index()
t2
```

```python
#using barplot to plot t2
sns.barplot(data = t2, x = 'class', y = 'count',color="r")
```

```python
#countplot of "class"
#argument called "hue"

sns.countplot(data=t,x="class",hue="sex")
```

```python
#violin plot

sns.violinplot(y=df["Height"],x=df["Gender"])
```

```python
#2 continuous variables seen in a scatter plot

sns.scatterplot(data=df,x="Height",y="Weight",hue="Gender")
```

```python
#old method

fig, ax = plt.subplots(2,2)
ax[0,1].plot(['sunday', 'monday'], [10, 27])
```

```python
#NEW method of creating subplots

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.scatterplot(data=df,x="Height",y="Weight",hue="Gender")
plt.xticks(rotation=90,size=10)

plt.subplot(1,3,2)
sns.boxplot(data=df,x="Gender",y="Weight")


plt.subplot(1,3,3)
sns.barplot(data=df,x="Gender",y="Height",estimator= np.min,ci=None)
plt.legend(['M','F'])
```
